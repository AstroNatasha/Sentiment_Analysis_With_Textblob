```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# Read the CSV file
csv_file = "Reviews.csv"
df_review = pd.read_csv(csv_file, nrows=100)  # Limiting to 100 rows

# Convert the "Text" column to string type
df_review['Text'] = df_review['Text'].astype(str)
```

```python
# Perform sentiment analysis
OldRange = (1. - (-1.))
NewRange = (5 - 1)  # Change the range to 1 to 5

# Apply sentiment analysis and scale to the new range
df_review['Sentiment'] = (((df_review['Text'].apply(lambda v: TextBlob(v).sentiment.polarity) - (-1.)) * NewRange) / OldRange) + 1.

# Round the scores to the nearest integer
df_review['Sentiment'] = df_review['Sentiment'].round().astype(int)
```

```python
# Visualize the sentiment distribution
plt.hist(df_review['Sentiment'], alpha=0.8, bins=range(1, 7), label="Sentiment")  # Adjust the bins accordingly
plt.legend()
plt.grid()
plt.ylabel('Frequency')
plt.xlabel('Rate')
plt.title('Sentiment Histogram - Review (Top 100 Rows)')
#plt.savefig('SentimentReview.png') optional
plt.show()
```

```python
# Visualize the actual rate distribution
plt.hist(df_review['Score'], alpha=0.8, bins=range(1, 7), label="Score")
plt.grid()
plt.ylabel('Frequency')
plt.xlabel('Rate')
plt.title('Sentiment Histogram - Actual rate (Top 100 Rows)')
#plt.savefig('nameoffigure.png') optional
plt.show()
```

```python
OldRange = (1. - (0.))
NewRange = (5 - 1)  # Change the range to 1 to 5

# Apply sentiment analysis and scale to the new range
df_review['Subjectivity'] = (((df_review['Text'].apply(lambda v: TextBlob(v).sentiment.subjectivity) - (0.)) * NewRange) / OldRange) + 1.

# Round the scores to the nearest integer
df_review['Subjectivity'] = df_review['Subjectivity'].round().astype(int)
```

```python
# Visualize the subjectivity distribution
plt.hist(df_review['Subjectivity'], alpha=0.8, bins=range(1, 7), label="Subjectivity")  # Adjust the bins accordingly
plt.legend()
plt.grid()
plt.ylabel('Frequency')
plt.xlabel('Rate')
plt.title('Subjectivity Histogram - Review (Top 100 Rows)')
#plt.savefig('SubjectivityReview.png') optional
plt.show()
```
